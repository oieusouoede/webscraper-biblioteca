```python
import requests
from bs4 import BeautifulSoup

class Busca:
    def __init__(self, cod):
        self.url_programa = 'https://dados.fflch.usp.br/programas/docentes/' + cod
        self.cod_fonte = requests.get(self.url_programa).text
        self.sopa_bonita = BeautifulSoup(self.cod_fonte, 'html.parser')
        self.texto = []

    def busca_artigos(u):
        pagina = requests.get(u).text
        sopa = BeautifulSoup(pagina, 'html.parser')
        a = []    
        for row in sopa.find_all('ul', attrs = {"id": "collapseArtigos"}):
            for i in row.find_all('li'):
                a.append(i.text)
                
        return a
    
    def busca_dados(self):
        for cada_linha in self.sopa_bonita.find_all('td', attrs = {'class': 'first-col'}):
            biscoito = cada_linha.a['href']
            url_docente = "https://dados.fflch.usp.br" + str(biscoito) + "&selection=artigos"
            nome_docente = cada_linha.span.text
            artigos = Busca.busca_artigos(url_docente)

            self.texto.append(nome_docente)
            self.texto.append(artigos)
            
cod = input('Insira o código do programa: ')

dados = Busca(cod)
dados.busca_dados()

dados.texto
```
